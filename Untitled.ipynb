```python
# Filter out the TIFF files
tif_files = [os.path.join(temp_dir, f) for f in dir_contents if f.endswith('.tif')]

# Create an empty list to store xarray datasets
datasets = []
```

```python
len(tif_files)
```

```python
# Open each TIFF file using xarray
for tif_file in tif_files:
    file_path = os.path.join(tif_file)
    dataset = xr.open_dataset(file_path)
    datasets.append(dataset)

# Now you have a list of xarray datasets, you can process them as needed
# For example, you can merge them into a single xarray dataset using xarray.concat()

```

```python
# Concatenate the list of datasets along the time dimension
combined_dataset = xr.concat(datasets, dim='time')
```
